```python
import matplotlib.pyplot as plt 
%matplotlib notebook

from sklearn.datasets import make_classification
Xc_2,_= make_classification(n_samples=200,
                                    n_features=2,
                                    n_informative=2,
                                    n_redundant=0,
                                    random_state=0,
                                    n_clusters_per_class=1,
                                    class_sep = 0.8)
plt.figure()
plt.scatter  (Xc_2[:,0], Xc_2[:,1])
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans

Xc_2
```

```python
kmeans = KMeans(4)
kmeans.fit(Xc_2)
```

```python
identified_clusters = kmeans.fit_predict(Xc_2)
identified_clusters
```

```python
x1, x2 = np.array(list((zip(*Xc_2))))
```

```python
plt.figure()
colors = np.array(['green','orange','grey', 'brown'])

plt.scatter(x1,x2, c= colors[identified_clusters])

centroids = kmeans.cluster_centers_  

# draw centroids 
for i, c in enumerate (kmeans.cluster_centers_):    
    plt.plot(c[0], c[1], marker = 'x', color= colors[i], markersize=14) 
    
plt.title ('sklearn kmeans')
```
