# Home task: Decision Trees


```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn import tree
import graphviz
```

```python
import os
os.environ["PATH"] += os.pathsep + r'C:\Program Files\Graphviz\bin'
```

## Brest cancer dataset


```python
cancer = load_breast_cancer()
X, y, labels, features = cancer.data, cancer.target, cancer.target_names, cancer.feature_names

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## DecisionTree


```python
print('DecisionTreeClassifier')
clf_decTree = DecisionTreeClassifier(
    criterion=  'entropy',
    random_state=20,        
    max_depth=3,    
).fit(X_train, y_train)

print("train accuracy= {:.3%}".format(clf_decTree.score (X_train, y_train)))
print("test accuracy= {:.3%}".format(clf_decTree.score (X_test, y_test)))

graph_viz = tree.export_graphviz(clf_decTree, out_file=None,feature_names=features, class_names=labels, filled = True)
graph = graphviz.Source(graph_viz)
graph.view(cleanup =True)
```

## RandomForest


```python
print("RandomForestClassifier")
clf_ranFor = RandomForestClassifier(
            random_state=20,
            n_estimators=10,      
        ).fit(X_train, y_train)

print("train accuracy= {:.3%}".format(clf_ranFor.score (X_train, y_train)))
print("test accuracy= {:.3%}".format(clf_ranFor.score (X_test, y_test)))
```

## Gradient Boosting Decision Trees (GBDT)


```python
print ('GradientBoostingClassifier')
clf_gradBoost = GradientBoostingClassifier(
    learning_rate=0.01,
    max_depth=3
).fit(X_train, y_train)

print("train accuracy= {:.3%}".format(clf_gradBoost.score (X_train, y_train)))
print("test accuracy= {:.3%}".format(clf_gradBoost.score (X_test, y_test)))

plt.figure ()
```

## XGBoost

```python
print ('XGBClassifier')
clf_xgb = XGBClassifier(use_label_encoder=False, eval_metric= 'logloss')
clf_xgb.fit(X_train, y_train)

print("train accuracy= {:.3%}".format(clf_xgb.score (X_train, y_train)))
print("test accuracy= {:.3%}".format(clf_xgb.score (X_test, y_test)))
```
